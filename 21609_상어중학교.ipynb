### 문제    
https://www.acmicpc.net/problem/21609   
_    
크기가 N×N인 격자, 격자 내 블록은 검은색 블록, 무지개 블록, 일반 블록이 있다.   
검은색 블록은 -1, 무지개 블록은 0으로 표현, 일반 블록은 M가지 색상이 있고, 색은 M이하의 자연수로 표현.
현 위치 기준 상,하,좌,우를 인접한 칸으로 본다.

**블록그룹**은 연결된 블록의 집합. 블록그룹의 조건은    
- 일반블록 1개 이상(일반블록 색 같아야함), 검정블록 포함 X, 무지개블록 개수 노상관.   
- 그룹 내 블록개수 2개 이상.      
- 그룹내 모든 칸은 서로 인접한 칸.

블록 그룹의 **기준블록**은 무지개 블록이 아닌 블록 중 행의 번호가 작은 블록 > 열의 번호가 작은 블록 

[오토플레이 - 아래과정 반복]
1. 크기가 가장 큰 **블록그룹** 찾기 > (여러개면) 무지개 블록수가 가장 많은 **블록그룹** > (여러개면) **기준블록** 행 번호 가장 큰 **블록그룹** > (여러개면) **기준블록** 열 번호 가장 큰 **블록그룹**   
2. 1에서 찾은 블록그룹의 블록 제거. 블록수^2을 점수로 획득   
3. 중력 작용     
4. 90도 반시계방향 회전    
5. 중력 작용  

#### 입력  
첫째 줄 :N M (격자 한 변의 크기 N, 색상의 개수 M) (1 ≤ N ≤ 20, 1 ≤ M ≤ 5)    
둘째 줄~:격자의 칸에 들어있는 블록의 정보 (1번 행 ~ N번 행) (입력으로 주어지는 칸의 정보는 -1, 0, M이하의 자연수)    

#### 출력  
첫째 줄에 획득한 점수의 합 출력    

#### 예제 
[input]   
5 3   
2 2 -1 3 1    
3 3 2 0 -1    
0 0 0 1 2   
-1 3 1 3 2    
0 3 2 2 1   

[output]    
77   

#### 조건 
- 시간제한 : 1초     
- 메모리제한 : 1024MB

```python
# 참고 코드 : https://blog.naver.com/jungs-note/223170042298

from collections import deque 

N, M = map(int, input().split())
blocks=[]
for _ in range(N):
    blocks.append(list(map(int, input().split())))
    
## 블록그룹 정보 반환 함수
def bfs(x, y):
    visited = []  # 그룹 내 블록 좌표 list 
    queue = deque()
    queue.append((x, y))
    cnt = 0  # 블록 그룹 크기
    rainbow_cnt = 0  # 무지개 블록 개수
    
    while queue:
        i, j = queue.popleft() ## queue에서 뽑은 블록(좌표)
        if (i, j) not in visited: # 그룹에 없는 좌표면 그룹에 넣음  
            visited.append((i, j))
        cnt += 1 
        
        if blocks[i][j] == 0: # (i,j) 위치에 무지개블록 존재시  
            rainbow_cnt += 1
        
        # 상하좌우 탐색 
        if i - 1 >= 0 and (i - 1, j) not in visited: # 블록이 존재하는지 & 그룹에 없는지 
            if blocks[i - 1][j] == blocks[x][y] or blocks[i - 1][j] == 0: # 색이 같거나 무지개블록이면 
                visited.append((i - 1, j))
                queue.append((i - 1, j))
        if i + 1 < N and (i + 1, j) not in visited: # 블록이 존재하는지 & 그룹에 없는지 
            if blocks[i + 1][j] == blocks[x][y] or blocks[i + 1][j] == 0: # 색이 같거나 무지개블록이면
                visited.append((i + 1, j))
                queue.append((i + 1, j))
        if j - 1 >= 0 and (i, j - 1) not in visited: # 블록이 존재하는지 & 그룹에 없는지 
            if blocks[i][j - 1] == blocks[x][y] or blocks[i][j - 1] == 0: # 색이 같거나 무지개블록이면
                visited.append((i, j - 1))
                queue.append((i, j - 1))
        if j + 1 < N and (i, j + 1) not in visited: # 블록이 존재하는지 & 그룹에 없는지 
            if blocks[i][j + 1] == blocks[x][y] or blocks[i][j + 1] == 0: # 색이 같거나 무지개블록이면
                visited.append((i, j + 1))
                queue.append((i, j + 1))
                
    # 기준 블록을 행, 열 번호가 가장 작은 블록으로 만드려고 정렬 (tuple 정렬*), 블록 수, 무지개블록 수
    return sorted(visited), cnt, rainbow_cnt


# 1. 크기가 가장 큰 블록그룹 찾기 함수 (블록그룹 정보 반환 함수 이용)
def find_max_group():
    visited = []  # 모든 방문 좌표 저장
    max_arr = []  # 블록 그룹 좌표 저장
    max_area = 1  # 블록 그룹 최대 크기
    max_rainbow = 0  # 무지개 블록 최대 개수
    
    for i in range(N):
        for j in range(N):
            # 이미 방문한 블록이면 skip
            if (i, j) in visited:
                continue
            
            # 빈 칸 / 검은색 블록 / 무지개 블록은 기준 블록으로 쓰지 않음
            if blocks[i][j] != "X" and blocks[i][j] != -1 and blocks[i][j] != 0:
                arr, area, rainbow = bfs(i, j) # 기준블록을 bfs 요소로 넣어서 블록그룹 정보 반환
                visited += arr 
                
                # 기존보다 크기가 더 큰 블록그룹 등장시 update 
                if area > max_area:
                    max_arr = arr
                    max_area = area
                    max_rainbow = rainbow
                    
                # (크기가 큰 블록 그룹이 여러개면) 포함된 무지개 블록의 수가 가장 많은 블록 그룹
                elif area == max_area and rainbow > max_rainbow:
                    max_arr = arr
                    max_rainbow = rainbow
                    
                # (무지개 블록의 수 같은 블록 그룹이 여러개면) 기준 블록의 행이 가장 큰 것을, 그것도 여러개이면 열이 가장 큰 것을 찾음 (bfs에서 arr 내부의 tuple이 정렬 되어있으므로 당연히 다음 arr의 행/열이 더 큼)
                elif area == max_area and rainbow == max_rainbow:
                    max_arr = arr
                    
    return max_arr, max_area

## 블록 하강시키는 함수
def dfs(i, j, color):
    if i + 1 < N and blocks[i + 1][j] == 'X': # 아랫칸이 빈칸인 경우 
        blocks[i][j] = 'X' 
        dfs(i + 1, j, color)
    else:
        blocks[i][j] = color
    return
    
# 3. 중력 작용 함수 (블록 하강시키는 함수 이용)
def gravity():
    for j in range(N):
        for i in range(N):
            # 맨 아래 칸부터
            if blocks[N - 1 - i][j] == -1: # 아래가 -1(검정블록)이면 하강불가 
                continue
            elif blocks[N - 1 - i][j] != 'X': # (하강이 가능한) 무지개블록 이거나 일반블록인 경우 
                dfs(N - 1 - i, j, blocks[N - 1 - i][j])      


### 오토 플레이 구현 ###
score = 0

while True:
    # 1. 블록그룹찾기
    max_group, b = find_max_group()

    # 블록 그룹이 존재하지 않으면 종료
    if b < 2:
        break

    # 2. 블록 제거 (제거한 블록값은 'X'라는 값으로 변경해줌)
    for (r, c) in max_group:
        blocks[r][c] = 'X'
    score += b ** 2 # 점수추가 

    # 3. 중력 작용
    gravity()

    # 4. 90도 반시계 방향으로 회전
    blocks = [x[::-1] for x in list(map(list, zip(*blocks[::-1])))[::-1]]

    # 5. 중력 작용
    gravity()

print(score)
```
