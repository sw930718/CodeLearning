```python
## 출처 : https://blog.naver.com/ongbbb/222721854171
from collections import deque
import copy

#전체는 dfs로 구성해서 모든 경우 고려해서 점수를 반환
#같은 번호의 차의 위치 찾기 - bfs로 찾기, 열이랑 사라지는 차 개수 저장해서  차 내려올 때 고려 
#최소 직사각형 찾기 
#점수 구하기
#자동차 위에서 내려오기

# input 
n = int(input())
arr = []

for _ in range(3*n):
    arr.append(list(map(int, input().split())))
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]
result = 0


# code 
def find_car(i,j,loc,visited,arr):
    temp = copy.deepcopy(arr[2*n:3*n])
    arr[2*n+i][j] = 0
    loc.append([i,j])
    q = deque()
    q.append([i,j,0])
    visited[i][j] = True
    
    while q:
        x, y, cnt = q.popleft()
        
        for d in range(4):
            nx = x + dx[d]
            ny = y + dy[d]
            if 0<=nx<n and 0<=ny < n and not visited[nx][ny]:
                
                if temp[x][y] == temp[nx][ny]:
                    arr[2*n + nx][ny] = 0
                    q.append([nx,ny,cnt+1])
                    visited[nx][ny] = True
                    loc.append([nx, ny])
    return loc[:]


def dfs(now, score, arr):
    global result
    if now == 3:
        result = max(result, score)  
        return result 

    arr = copy.deepcopy(arr)
    visited = [[False] * n for _ in range(n)]   
    # save_loc = []
    for i in range(n):
        for j in range(n):
            pre_arr = copy.deepcopy(arr)
            if visited[i][j] or arr[2*n+i][j] == 0: continue

            car_loc = find_car(i,j,[],visited,arr)
            car_cnt = len(car_loc)



            #직사각형 찾기, 점수 구하기
            col_info = [] #[col_min, col_max]

            row = sorted(car_loc, key = lambda x:-x[0]) #행번호 찾기
            row_min, row_max = row[-1][0], row[0][0]
        
            col = sorted(car_loc, key = lambda x: -x[1]) #열번호 찾기
            col_min, col_max = col[-1][1], col[0][1]

            col_info.append([col_min, col_max])
            temp_score = car_cnt + (row_max - row_min + 1)*(col_max - col_min+1)



            #차 내려오기
            for y in range(col_min, col_max+1):
                for x in range(2*n-1, -1, -1):
                    if arr[x+1][y] != 0:
                        continue
                    nx = x
                    while True: 
                        if nx == 3*n-1:
                            break
        
                        if nx+1 < 3*n and arr[nx+1][y] == 0:
                            nx = nx + 1

                        else: break

                    arr[nx][y] = arr[x][y]
                    arr[x][y] = 0

            dfs(now+1, score + temp_score, arr)
            arr = copy.deepcopy(pre_arr)

dfs(0, 0, arr)
print(result)
```
