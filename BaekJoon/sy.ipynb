```python
"""
최소 x번만에 도착하는 방법을 이야기하므로.. 
최단경로 문제와 유사하다고 판단. -> BFS로 해결

규칙
1. 사다리 만나면 x칸에서 y칸으로 바로 이동 가능
2. 뱀 만나면 u칸에서 v칸으로 빠꾸
3. 시작은 항상 1번칸에서 시작

생각
1. 사다리 타면 무조건 이득
2. 뱀 무조건 피해야됨
3. 사다리만 타고 쭉쭉 올라가면 끝
4. 근데 어떤 사다리를 고르냐? 가 되게 이슈
5. 일단 사다리 중 최대값을 구하고, 그 사다리를 타기 위해 이동
6. 만약 이동하다가 최댓값 사다리 시작점 > 다른 사다리 끝점 이 있으면 그거 이용 (그게 6보다 크면!)
"""
예상 예외사항 
1. 61->92, 12->90 중에 이 코드는 61->92로 선택됨. 최단은 12->90임
2. 사다리 타고 내렸는데, 앞에 뱀이 깔려있는 경우 ? 
3. 나온 이야기중에 뱀을 타야 더 짧게 올라갈 수 있는 경우가 있었음 -> so, 뱀 고려는 필수 

from collections import deque

N, M = map(int, input().split())
xys = [tuple(map(int, input().split())) for _ in range(N)]
uvs = [tuple(map(int, input().split())) for _ in range(M)]


q = deque(sorted(xys, key= lambda x: x[1], reverse=True)) 

max_start, max_end = q.popleft()

will_be_used_ladders = []
will_be_used_ladders.append((max_start, max_end))
for idx, (start, end) in enumerate(q):
    if end <= max_start:
        will_be_used_ladders.append((start, end))
        max_start, max_end = start, end

cnt = 0
now = 1
tmp_plus = 6
for ladder_start, ladder_end in will_be_used_ladders:
    if now < ladder_start:
        move, mod = divmod(ladder_start, 6)
        now = ladder_end
        cnt += move + 1

if now == 100:
    print(cnt)
else:
    move, mod = divmod(100-now, 6)
    cnt += move 
    print(cnt)
            
#########################################################################
from collections import deque

def snake_and_ladder(N, M, xys, uvs):
    ladder_starts = {}  # 사다리 시작점을 저장하는 딕셔너리
    snake_heads = {}  # 뱀 시작점을 저장하는 딕셔너리

    for x, y in xys:
        ladder_starts[x] = y

    for u, v in uvs:
        snake_heads[u] = v

    def bfs():
        visited = [False] * 101
        queue = deque([(1, 0)])  # 시작 위치와 주사위 굴린 횟수

        while queue:
            position, count = queue.popleft()

            for dice in range(1, 7):
                next_position = position + dice

                if next_position in ladder_starts:
                    next_position = ladder_starts[next_position]

                if next_position in snake_heads:
                    next_position = snake_heads[next_position]

                if next_position == 100:
                    return count + 1

                if not visited[next_position]:
                    visited[next_position] = True
                    queue.append((next_position, count + 1))

    return bfs()

N, M = map(int, input().split())
xys = [tuple(map(int, input().split())) for _ in range(N)]
uvs = [tuple(map(int, input().split())) for _ in range(M)]

result = snake_and_ladder(N, M, xys, uvs)
print(result)
```
